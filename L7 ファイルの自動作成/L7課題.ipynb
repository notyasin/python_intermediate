# 練習課題: シンプルな自動テキストファイル作成

## 説明

この練習課題では、Pythonスクリプトを作成して、シンプルな内容のテキストファイルを自動的に生成する練習を行います。課題は簡単に進めるために簡素化されています。

以下の手順に従って課題を完了してください：

### ステップ1: セットアップ

1. Jupyter Notebookの蘭を一個を開きます。

### ステップ2: 練習課題のタスク

次のタスクを実行するスクリプトを作成してください。各テキストファイルには以下の内容が含まれるようにします：

1. ファイル番号を示すヘッダー。
2. シンプルなメッセージ。

以下の手順に従って課題を完了してください：

1. 作成するファイルの数を設定します（`num_files = 3`）。
2. ファイル番号の範囲を反復する `for` ループを作成します。
3. ループ内で以下を行います：
   - ファイル番号を使用してファイル名を生成します。
   - ファイルの内容を生成します：
     - ファイル番号を示すヘッダー。
     - "こんにちは、これはファイル番号Xです。" といったシンプルなメッセージ。
   - コンテンツをファイルに書き込みます。
   - 各作成されたファイルに対して成功メッセージを表示します。

### ステップ3: 実行とテスト

1. スクリプトを実行します。
2. スクリプトが配置されているディレクトリを確認し、新しく作成されたテキストファイルを確認します。
3. 作成されたテキストファイルの1つを開き、期待される内容が含まれていることを確認します。

## 例としての出力

次は、生成されたテキストファイルの内容の一例です：

```
ファイル: file_1.txt

こんにちは、これはファイル番号1です。
```
